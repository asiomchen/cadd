```python
import pandas as pd
import numpy as np
from centroid import *
import os, sys
```

```python
DATA_PATH = './results/docked_smina_auto_128'
!ls ./results/docked_smina_auto_128
```

```python
i = 0
data = {'name': [],
        'pose': [],
        'centroid':[]}
for file in os.listdir(DATA_PATH):
    i+=1
    print(i)
    file_path = DATA_PATH + '/' + file
    fff = from_sdf_to_xyz(file_path)
    pose = 1
    for x in centroid(xyz_to_numpy(fff)):
        data['name'].append(file)
        data['pose'].append(pose)
        data['centroid'].append(x)
        print(file)
        print(f'Pose: {pose}')
        print(x)
        pose+=1
    os.remove(fff)
print(data)
```

```python
df = pd.DataFrame.from_dict(data)
df.head()
```

```python
df['centroid_x'] = df['centroid'].apply(lambda x: x[0])
df['centroid_y'] = df['centroid'].apply(lambda x: x[1])
df['centroid_z'] = df['centroid'].apply(lambda x: x[2])
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot_points(df):
    """
    Plots clusters made from vectorized procedure in 3D
    param:matrix - np.ndarray of vectors to plot
    param:model - sklearn model to predict labels
    return: pd.Dataframe
    """

    x_axis = [o for o in df['centroid_x']]
    y_axis = [o for o in df['centroid_y']]
    z_axis = [o for o in df['centroid_z']]
    df_dict = {'id':df.name, 'x': x_axis, 'y': y_axis, 'z':z_axis}
    df_plot = pd.DataFrame(df_dict)

    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111, projection='3d')
    for i in df_plot.id.unique():
        i_cluster = df_plot[df_plot.id == i]
        x = i_cluster['x']
        y = i_cluster['y']
        z = i_cluster['z']
        s = 40
        if 'CHEMBL145.' in i or 'CHEMBL6.' in i or 'CHEMBL1316' in i:
            print('aaaaa')
            s = 150
        ax.scatter3D(x,y,z, s=s, label=i, alpha=1)
    ax.legend()
    plt.show()

    return df_plot
```

```python
cleaned['size'] = 20 * (cleaned.value - cleaned.value.max()) ** 2 + 20
```

```python
def plot_points(df):
    """
    Plots clusters made from vectorized procedure in 3D
    param:matrix - np.ndarray of vectors to plot
    param:model - sklearn model to predict labels
    return: pd.Dataframe
    """

    x_axis = [o for o in df['centroid_x']]
    y_axis = [o for o in df['centroid_y']]
    z_axis = [o for o in df['centroid_z']]
    size = [o for o in df['size']]
    df_dict = {'id':df.name, 'x': x_axis, 'y': y_axis, 'z':z_axis, 'size': size}
    df_plot = pd.DataFrame(df_dict)

    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111, projection='3d')
    for i in df_plot.id.unique():
        i_cluster = df_plot[df_plot.id == i]
        x = i_cluster['x']
        y = i_cluster['y']
        z = i_cluster['z']
        s = i_cluster['size']
        ax.scatter3D(x,y,z, s=s, label=i, alpha=0.2)
    #ax.legend()
    plt.show()

    return df_plot
```

```python
ploted = plot_points(cleaned)
```

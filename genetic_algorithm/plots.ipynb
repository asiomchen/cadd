```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set_style("white")
```

```python
firts_gen = df[df.gen == 0]
last_gen = df[df.gen == 49]
```

```python
firts_gen.lig.value_counts()
```

```python
firts_gen_freq = dict(firts_gen['lig'].value_counts())
```

```python
firts_gen_freq
```

```python
firts_gen['lig_freq'] = firts_gen['lig']
firts_gen['lig_freq'] = firts_gen['lig_freq'].map(firts_gen_freq)
```

```python
firts_gen
```

```python
last_gen_freq = dict(last_gen['lig'].value_counts())
```

```python
last_gen['lig_freq'] = last_gen['lig']
last_gen['lig_freq'] = last_gen['lig_freq'].map(last_gen_freq)
```

```python
last_gen
```

```python
merged = pd.concat([firts_gen, last_gen])
present_idx = merged[(merged.lig_freq > 1) & (merged.gen == 49)].lig
lig_plot = merged[merged.lig.isin(present_idx)]
```

```python
sns.countplot(data=lig_plot, x='lig', hue='gen')
```

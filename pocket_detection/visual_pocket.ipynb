```python
import pandas as pd
import numpy as np
import sys,os
sys.path.insert(0,'../')
from extract_scores import extract_scores_from_dir
```

```python
base_name = '../results/ares_qvina/docked_pocket_'
temp_df = None
all_results = None
for e in [16,32,64,128,192]:
    res_dir = base_name + str(e)
    scores = extract_scores_from_dir(res_dir)
    temp_df = pd.DataFrame.from_dict(scores, orient='columns')
    temp_df['e'] = e
    temp_df.reset_index(inplace=True)
    if all_results is None:
        all_results = temp_df.copy()
    else:
       all_results = pd.concat([all_results, temp_df], axis=0)
    print(temp_df.head())
```

```python
all_results['centroid_x'] = all_results['centroid'].apply(lambda x: x[0])
all_results['centroid_y'] = all_results['centroid'].apply(lambda x: x[1])
all_results['centroid_z'] = all_results['centroid'].apply(lambda x: x[2])
```

```python
def plot_points(df):
    """
    Plots clusters made from vectorized procedure in 3D
    param:matrix - np.ndarray of vectors to plot
    param:model - sklearn model to predict labels
    return: pd.Dataframe
    """

    x_axis = [o for o in df['centroid_x']]
    y_axis = [o for o in df['centroid_y']]
    z_axis = [o for o in df['centroid_z']]
    df_dict = {'id':df.name, 'x': x_axis, 'y': y_axis, 'z':z_axis}
    df_plot = pd.DataFrame(df_dict)

    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111, projection='3d')
    for i in df_plot.id.unique():
        i_cluster = df_plot[df_plot.id == i]
        x = i_cluster['x']
        y = i_cluster['y']
        z = i_cluster['z']
        ax.scatter3D(x,y,z, s=40, label=i, alpha=1)
    ax.legend()
    plt.show()

    return df_plot
```

```python
to_cluster = plot_points(all_results)
```

```python
from sklearn.cluster import KMeans
```

```python
len(to_cluster)
```

```python
inertias = []
n_clusters = []
for n in range(1, 13, 1):
    model = KMeans(n_clusters=n)
    model.fit(to_cluster.drop('id', axis=1))
    inertias.append(model.inertia_)
    n_clusters.append(n)
sns.lineplot(x=n_clusters, y=inertias)
```

```python
labels = KMeans(n_clusters=10).fit(to_cluster).labels_
```

```python
from sklearn.decomposition import PCA

coords = to_cluster[['x','y', 'z']]
pca = PCA(n_components=2)
coords
```

```python
coords = pca.fit_transform(coords)
plt.scatter(coords[:,0], coords[:,1], alpha=0.5)
plt.show()
```

```python
model_2 = KMeans(n_clusters=10).fit(coords)
pca_labels = model_2.labels_
```

```python
sns.scatterplot(x=coords[:,0], y=coords[:,1], alpha=0.5, hue=pca_labels, palette=sns.color_palette())
plt.show()
```

```python
all_results['pocket'] = pca_labels
```

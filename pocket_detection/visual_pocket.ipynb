```python
def plot_points(df, legend=False):
    """
    Plots clusters made from vectorized procedure in 3D
    param:matrix - np.ndarray of vectors to plot
    param:model - sklearn model to predict labels
    return: pd.Dataframe
    """
    sns.set_style("white")

    x_axis = [o for o in df['centroid_x']]
    y_axis = [o for o in df['centroid_y']]
    z_axis = [o for o in df['centroid_z']]
    df_dict = {'id':df.name, 'x': x_axis, 'y': y_axis, 'z':z_axis}
    df_plot = pd.DataFrame(df_dict)

    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111, projection='3d')
    for i in df_plot.id.unique():
        i_cluster = df_plot[df_plot.id == i]
        x = i_cluster['x']
        y = i_cluster['y']
        z = i_cluster['z']
        ax.scatter3D(x,y,z, s=40, label=i, alpha=1)
    if legend:
        ax.legend()
    plt.show()

    return df_plot
```

```python
def elbow_plot(method=None, data=None, lims=(5,15)):
    inertias = []
    n_clusters = []
    for n in range(lims[0], lims[1] + 1):
        model = method(n_clusters=n)
        model.fit(data)
        inertias.append(model.inertia_)
        n_clusters.append(n)
    sns.set_style("white")
    sns.lineplot(x=n_clusters, y=inertias)
    plt.title('Elbow plot')
    plt.ylabel('Inertia')
    plt.xlabel('Number of clusters')
    plt.xlim(lims[0], lims[1])
```

```python
def plot_clusters(df, cluster='labels', alpha=0.5):
    """
    Plots clusters made from vectorized procedure in 3D
    param:matrix - np.ndarray of vectors to plot
    param:model - sklearn model to predict labels
    return: pd.Dataframe
    """
    # color_labels = df[hue].unique()
    # rgb_values = sns.color_palette("Set2", len(color_labels))
    # color_map = dict(zip(color_labels, rgb_values))

    x_axis = [o for o in df['centroid_x']]
    y_axis = [o for o in df['centroid_y']]
    z_axis = [o for o in df['centroid_z']]


    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111, projection='3d')
    for i in df[cluster].unique():
        i_cluster = df[df[cluster] == i]
        x = i_cluster['centroid_x']
        y = i_cluster['centroid_y']
        z = i_cluster['centroid_z']
        ax.scatter3D(x,y,z, s=40, label=i, alpha=alpha)
    ax.legend()
    plt.show()
```

```python
def get_cluster_centroid(df, labels='DBSCAN'):
    out = []
    for i in df[labels].unique():
        subset = df[df[labels] == i]
        out.append([subset.centroid_x.mean(), subset.centroid_y.mean(), subset.centroid_z.mean()])
    return out, df[labels].unique()
        
```

```python
no_noise = all_results[all_results.DBSCAN != -1]
sns.scatterplot(x=no_noise.centroid_x, y=no_noise.centroid_y, hue=no_noise.DBSCAN, palette=sns.color_palette(n_colors=20), alpha=0.3)
plt.title('Clustered centroids - mean')
energies = no_noise.groupby('DBSCAN')['score'].mean().values
energies_median = no_noise.groupby('DBSCAN')['score'].median().values
cluster_centers, indicies = get_cluster_centroid(no_noise)
for centroid, energy, median, idx in zip(cluster_centers, energies, energies_median, indicies):
    if idx == -1:
        print('Skipped noisy samples')
        continue
    energy_str = str(energy)[:6]
    energy_median = str(median)[:6]
    x = centroid[0]
    y = centroid[1]
    #plt.text(x + 0.5, y, energy_str, horizontalalignment='left', size='medium', color='black', weight='semibold')
    plt.text(x + 3, y - 2, energy_str, horizontalalignment='left', size='medium', color='black', weight='bold')
```

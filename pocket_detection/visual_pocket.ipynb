```python
all_results['centroid_x'] = all_results['centroid'].apply(lambda x: x.x)
all_results['centroid_y'] = all_results['centroid'].apply(lambda x: x.y)
all_results['centroid_z'] = all_results['centroid'].apply(lambda x: x.z)
```

```python
def plot_points(df, legend=False):
    """
    Plots clusters made from vectorized procedure in 3D
    param:matrix - np.ndarray of vectors to plot
    param:model - sklearn model to predict labels
    return: pd.Dataframe
    """

    x_axis = [o for o in df['centroid_x']]
    y_axis = [o for o in df['centroid_y']]
    z_axis = [o for o in df['centroid_z']]
    df_dict = {'id':df.name, 'x': x_axis, 'y': y_axis, 'z':z_axis}
    df_plot = pd.DataFrame(df_dict)

    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111, projection='3d')
    for i in df_plot.id.unique():
        i_cluster = df_plot[df_plot.id == i]
        x = i_cluster['x']
        y = i_cluster['y']
        z = i_cluster['z']
        ax.scatter3D(x,y,z, s=40, label=i, alpha=1)
    if legend:
        ax.legend()
    plt.show()

    return df_plot
```

```python
to_cluster = plot_points(all_results).drop('id', axis=1)
```

```python
from sklearn.cluster import KMeans
```

```python
len(to_cluster)
```

```python
inertias = []
n_clusters = []
for n in range(4, 15, 1):
    model = KMeans(n_clusters=n)
    model.fit(to_cluster)
    inertias.append(model.inertia_)
    n_clusters.append(n)
sns.lineplot(x=n_clusters, y=inertias)
```

```python
def elbow_plot(method=None, data=None, lims=(5,15)):
    inertias = []
    n_clusters = []
    for n in range(lims[0], lims[1]):
        model = method(n_clusters=n)
        model.fit(data)
        inertias.append(model.inertia_)
        n_clusters.append(n)
    sns.lineplot(x=n_clusters, y=inertias)
```

```python
elbow_plot(KMeans, to_cluster, (1,20))
```

```python
def plot_clusters(df, cluster='labels', alpha=0.5):
    """
    Plots clusters made from vectorized procedure in 3D
    param:matrix - np.ndarray of vectors to plot
    param:model - sklearn model to predict labels
    return: pd.Dataframe
    """
    # color_labels = df[hue].unique()
    # rgb_values = sns.color_palette("Set2", len(color_labels))
    # color_map = dict(zip(color_labels, rgb_values))

    x_axis = [o for o in df['centroid_x']]
    y_axis = [o for o in df['centroid_y']]
    z_axis = [o for o in df['centroid_z']]


    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111, projection='3d')
    for i in df[cluster].unique():
        i_cluster = df[df[cluster] == i]
        x = i_cluster['centroid_x']
        y = i_cluster['centroid_y']
        z = i_cluster['centroid_z']
        ax.scatter3D(x,y,z, s=40, label=i, alpha=alpha)
    ax.legend()
    plt.show()
```

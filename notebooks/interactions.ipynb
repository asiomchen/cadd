```python
from rdkit import Chem

#load protein
prot = mda.Universe('./master/inter/cleaned_3.pdb', guess_bonds=True, vdwradii={"H": 0.95, 'Co':1.92})
print(mda.topology.guessers.tables.vdwradii)
# prot = plf.Molecule.from_mda(prot)
# prot.n_residues
```

```python
for atom in prot.atoms:
    # find hydrogens with more than one bond
    if atom.element == "H" and len(atom.bonds) > 1:
        neighbours = set([i for i in atom.bonds.indices.flatten() if i != atom.index])
        for i in neighbours:
            neighbour = prot.atoms[i]
            # find neighbour to H from different residue
            if neighbour.resid != atom.resid:
                print(f"{atom.resname}{atom.resid}:{atom.name}",
                      f"{neighbour.resname}{neighbour.resid}:{neighbour.name}")
```

```python
prot = plf.Molecule.from_mda(prot)
prot.n_residues
```

```python
def plot_interaction(ligand_file, protein_file=prot, frame=0):
    lig_suppl = list(plf.sdf_supplier(ligand_file))
    fp = plf.Fingerprint()
    fp.run_from_iterable(lig_suppl, protein_file)
    results_df = fp.to_dataframe(return_atoms=True)
    net = LigNetwork.from_ifp(results_df,lig_suppl[0],kind="frame", frame=frame,rotation=270)
    return net
```
